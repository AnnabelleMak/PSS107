```python
## how to generate 9x9 table?
# print a, b, ...The comma(,) means concatenation. 
# Another concatenation is PLUS(+), only workable when the data is "string" type
for i in range(1, 10):
    print i
for i in range(1, 10):
    print "%d* = "%i

for i in range(1, 10):
    for j in range(1, 10):
        print "%d*%d=%d "%(i, j, i*j),
```
