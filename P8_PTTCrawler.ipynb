````python
pre = 'https://www.ptt.cc'
links = []
for i in range(2904, 2904-10, -1):
    url = 'https://www.ptt.cc/bbs/Boy-Girl/index%d.html'%i
    results = requests.get(url)
    soup = BeautifulSoup(results.text)
    print type(soup.find_all(class_='r-ent'))
    for link in soup.find_all(class_='r-ent'):
        print '-'*50
#         print link.find(class_='nrec').text
#         print link.find(class_='date').text.strip()
#         print link.find(class_='author').text.strip()
        print link.find(class_='title').text.strip()
        try:
            print pre + link.find(class_='title').a['href']
            links.append(pre + link.find(class_='title').a['href'])
        except:
            pass
print len(links)
````

````python
from bs4 import NavigableString
all_post = []
for link in links:
    print link
    results = requests.get(link)
    soup = BeautifulSoup(results.text)
    metas = soup.find_all(class_='article-meta-value')
    author = metas[0].text
    title = metas[-2].text
    timestamp = metas[-1].text
    print author, title, timestamp
    
    content = ''
    for text in soup.find(id='main-content'):
        if isinstance(text, NavigableString):
            content += text.strip()
    
    comments = []
    for push in soup.find_all(class_="push"):
        push_userid = push.find(class_='push-userid').text
        push_content = push.find(class_='push-content').text
        push_ipdatetime = push.find(class_='push-ipdatetime').text

        #print push_userid, push_content, push_ipdatetime
        comments.append({'userid':push_userid, 
                         'content':push_content, 
                         'timestamp':push_ipdatetime})
    all_post.append({'author':author, 
                     'link':link,
                      'title':title, 
                      'timestamp':timestamp, 
                      'content':content, 
                      'comments':comments}) 
````

````python
import json
json.dump(all_post, open('ptt.json', 'w'))
````

````python
from bs4 import NavigableString

post_list = []
for link in links:
    
#     print link
    results = requests.get(link)
    soup = BeautifulSoup(results.text)
    meta = soup.find(id='main-content').find_all(class_='article-meta-value')

    content = ''
    for text in soup.find(id='main-content'):
        if isinstance(text, NavigableString):
            content += text.strip()

    comments = []
    for comment in soup.find_all(class_='push'):
        comments.append({'tag':comment.find(class_='push-tag').text, 
                                  'user':comment.find(class_='push-userid').text, 
                                  'content':comment.find(class_='push-content').text, 
                                  'time':comment.find(class_='push-ipdatetime').text.strip()})
    try:
        _post = {'author':meta[0].text, 
                 'link':link,
                 'board':meta[1].text, 
                 'title':meta[2].text, 
                 'time':meta[3].text, 
                 'content':content, 
                 'comments':comments}
        post_list.append(_post)
    except:
        print meta
print len(post_list)    
````
