# Read CSV

當我規劃要讀進一個csv檔時，我可能有三種做法：
1. 全部靠自己手動讀、切行、遇到逗號進行分隔。
2. 利用csv.reader()函式幫我把文字資料轉換成python物件
3. 利用pandas讀檔

```python
all_list = []
with open("data/14196_drug_adv.csv", "r") as f:
    for line in f.readlines():
        row = line.split(",")
        all_list.append(row)
#         print(len(row))
print(all_list[1][-2])
```

## Read by csv.reader()

```python
import csv
with open("data/14196_drug_adv.csv") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    data_list  = list(csv_reader)
print(data_list[0])
print(data_list[1])

# for row in data_list: # never do this if you are not sure that the number of lines
#     print(row)
print(data_list[1][-2])
print(len(data_list))
```

## Read by pandas data.frame

```python
import pandas as pd
df = pd.read_csv('data/14196_drug_adv.csv', error_bad_lines=False)
```

```python
# list(df["刊播媒體"])
df.刊播媒體類別
type_dict = {}
for item in list(df.刊播媒體類別):
    type_dict[item] = 0
for item in list(df.刊播媒體類別):
    type_dict[item] += 1
print(type_dict)
```
